```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
football = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
#for shrinking the rows
football.head(5)
```

```python
left= football.loc[football['Preferred Foot']=='Left'].count()[0]
right= football.loc[football['Preferred Foot']=='Right'].count()[0]
labels=['left','right']
plt.pie([left,right],labels=labels)
plt.show()
```

```python
one=50
two=50
plt.title("My Piechart")
colors=['#bbbbcb','#aacaaa']
labels=['one','two']
plt.pie([one,two], labels=labels, colors=colors, autopct='%0.0f %%')
plt.show()
```

```python
football.Weight= [int(x.strip('lbs')) if type(x)==str else x for x in football.Weight]
light=football.loc[(football.Weight<125)].count()[0]
medium_light= football.loc[(football.Weight>125) & (football.Weight<150)].count()[0]
medium = football.loc[(football.Weight>150) & (football.Weight<175)].count()[0]
medium_heavy= football.loc[(football.Weight>175) & (football.Weight<200)].count()[0]
heavy= football.loc[(football.Weight>200)].count()[0]
weight=[light,medium_light,medium,medium_heavy,heavy]
labels=['<125','125-150','150-175','175-200','>200']
explode= (0.4,0.2,0.0,0.0,0.4)
plt.pie(weight, labels=labels, explode=explode)
plt.title("Players Weights")
plt.show()
```
