```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
gas = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/gas_prices.csv')
plt.figure(figsize=(8,5))
plt.title("Gas Prices overtime" ,fontdict={'fontweight':'bold', 'fontsize':22})
plt.plot(gas.Year,gas.USA, 'rx-')
plt.plot(gas.Year,gas.Canada, 'bx-')
plt.plot(gas.Year,gas['South Korea'], 'yx-')
plt.xlabel('Countries',fontdict={'fontweight':'bold', 'fontsize':16})
plt.ylabel('Gas Prices',fontdict={'fontweight':'bold', 'fontsize':16})
plt.xticks(gas.Year[::3]) # here we are passing all the years form starting to ending with skipping 3 years in between
plt.show()
plt.savefig('Gases_Prices.png',dpi=300)
```

```python
#alternate method for the graph
selected_countries=['USA','Canada','South Korea']
for country in gas:
  if country in selected_countries:
    plt.plot(gas.Year,gas[country], marker='.')
```

```python
football = pd.read_csv('https://raw.githubusercontent.com/KeithGalli/matplotlib_tutorial/master/fifa_data.csv')
#for shrinking the rows
football.head(5)
```

```python
# Histograms
bins=[40,50,60,70,80,90]
plt.hist(football.Overall, bins=bins, color='#abcdef')
plt.xticks(bins)
plt.title("Distribution of Football players",fontdict={'fontweight':'bold', 'fontsize':16})
plt.xlabel("Skill Level")
plt.ylabel("No.of Players")
plt.figure(figsize=(8,5))
plt.show()
```
