```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
#Basic Graph
plt.plot([1,2,3],[2,4,6])
plt.show()     #to remove the unneccesary line in the graph
```

```python
#Basic Graph
x=[0,1,2,3,4]
y=[0,2,4,6,8]
plt.title('My First Graph',fontdict={'fontname':'Lucida Sans Unicode','fontsize':15})
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([0,1,2,3,4])
plt.yticks([0,2,4,6,8])
plt.plot(x,y, label='2x', linewidth=2, color='red', marker='.',linestyle='--', markersize=10, markeredgecolor='blue')
plt.show()
```

```python
x=[1,2,3,4]
y=[2,4,6,8]
plt.plot(x,y,'r^--') #shortcut: '[color][marker][line]'
plt.show()
```

```python
x2=np.arange(0,4,0.5) #x=[0, 0.5, 1, 1.5, 2, 2.5, 3, 3.5]
plt.plot(x2[:6],x2[:6]**2, 'b--')
plt.plot(x2[5:],x2[5:]**2,'b')
plt.figure(figsize=(1,1),dpi=300) #resizing the graph
plt.show()
```
